```python
# Linear graph
 
import matplotlib.pyplot as plt

plt.title("Basic Linear Graph")

plt.plot([1, 2, 3], [2, 4, 6])

plt.show()
```

```python
# Graph to plot between  distance and time

distance = list(range(2, 11, 2))

time = list(range(1,6))

#d = 2 * t

plt.plot(time, distance, 'go') # gives green points in a graph

plt.title("Distance vs Time") # returnsthe graph with title on top

plt.grid() # grid function gives the graph in grid model

plt.xlabel("Time") # gives the name x-axis

plt.ylabel("Distance") # gives tha name to  y-axis

plt.show()

plt.savefig("DvsT") # saves graph in the current path
```

```python
# Non Linear graph

# velocity = accleration * time

t = list(range(1, 11, 2))

a = list(range(1, len(t)+1))

v = [ a[i] * i for i in range(len(t))]

plt.xlabel("Time")

plt.ylabel("Velocity")

plt.plot(t, v)
plt.grid()
```

```python
# Graph b/w number and square of a number

n = list(range(10))

squares = [i * i for i in n]

plt.plot(n, squares)

plt.title("NumbersVSSquares")

plt.grid()

plt.show()
```

```python
# Voltage - Time graph

import numpy as np

time = [i/10 for i in range(1, 21)]

voltage = [1+ np.sin(2 *np.pi * i) for i in t]

plt.title('Time vs Voltage')

plt.plot(time, voltage, 'go')

plt.grid()

plt.savefig('images/Voltage-Time.png')

plt.show()
```

```python
# Create graph plot income of arkansas from 2005 to 20013

import pandas as pd

df = pd.read_csv("DataFiles/Income.csv")

years = df.columns[2:]

incomeArkansis = df.values[3, 2:]

plt.title('Income of Arkansis')

plt.xlabel('Year')

plt.ylabel('Income')

plt.plot(years, incomeArkansis)

plt.grid()

plt.savefig("images/incomeArkansas.png")

plt.show()
```

```python
# Income of all states from 2005 to 2013

averages = [df[i].mean() for i in years]

plt.plot(years, averages)

plt.title('Average income of all states ')

plt.xlabel('Years')

plt.ylabel('Income')

plt.show()
```
